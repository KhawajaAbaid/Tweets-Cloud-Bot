```python
#load up the mask
from PIL import Image
mask = np.array(Image.open(BASE_PATH / "data/twitter_mask_sq.png"))
```

```python
#load up the gradient
from wordcloud import ImageColorGenerator
gradient = ImageColorGenerator(np.array(Image.open(BASE_PATH / "data/twitter_cloud_gradient_sq.png")))
```

```python
wordcloud_generator = WordCloud(background_color="black", mask=mask, width=1000, height=1000, color_func=gradient)
wordcloud_img  = wordcloud_generator.generate(" ".join(sample_words))
wordcloud_img.to_file(BASE_PATH / "data/wordcloud_sq_ed_1x.png")
plt.imshow(wordcloud_img, interpolation='bilinear')
plt.axis("off")
# plt.figure()
# plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
# plt.axis("off")
plt.show()
```

```python
wordcloud_generator = WordCloud(background_color="white", mask=mask, width=1000, height=1000, color_func=gradient)
wordcloud_img  = wordcloud_generator.generate(" ".join(sample_words))
wordcloud_img.to_file(BASE_PATH / "data/wordcloud_sq_white.png")
plt.imshow(wordcloud_img, interpolation='bilinear')
plt.axis("off")
# plt.figure()
# plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
# plt.axis("off")
plt.show()
```

```python
# get the sketchy font
font_path = BASE_PATH / "fonts/CabinSketch-Regular.ttf"
```

```python
wordcloud_generator = WordCloud(background_color="white", mask=mask, width=1000, height=1000, font_path=font_path.as_posix())
wordcloud_img  = wordcloud_generator.generate(" ".join(sample_words))
# wordcloud_img.to_file(BASE_PATH / "data/wordcloud_sketch_with_info.png")
plt.text(450, 950, "Generated by TweetsCloud", fontsize=7, color="red")
plt.imshow(wordcloud_img, interpolation='bilinear')
plt.axis("off")
plt.savefig("wpltimsavehehehaha.png", dpi=300)
plt.show()
```
